```python
# Q4
print('Please enter a number')
my_number = input()
my_num = int(my_number)
start = 1
print(f'number | squared | cubed\n------ | ------- | -----')
while start < my_num:
    print(f'{start}      | {start ** 2}       | {start** 3}')
    start += 1
print('would you like to pick another number? (Y/N)') 
prompt = input()
while prompt.lower() == 'y':
    print('Please enter another number')
    my_next_num = input()
    my_num = int(my_next_num)
    start = 1
    print(f'number | squared | cubed\n------ | ------- | -----')
    while start < my_num:
        print(f'{start}      | {start ** 2}       | {start** 3}')
        start += 1
    print('would you like to pick another number? (Y/N)') 
    prompt = input()
else:
    print('Thank you!')
```

```python
# Q5

print(' Please enter your number grade into the letter grade converter.')
num_grade = input()
grade = int(num_grade)
if grade < 60:
    print(" Letter grade is an 'F'!")
elif grade > 59 and grade < 70:
    print(" Letter grade is a 'D'!")
elif grade > 69 and grade < 80:
    print(" Letter grade is a 'C'!")
elif grade > 79 and grade < 90:
    print(" Letter grade is a 'B'!")
else:
    print("Congratulations! Your letter grade is an 'A'!")
print('would you like to enter another grade into the converter? (Y/N)')
prompt = input()
if prompt.lower() == 'y':
    print(' Please enter your number grade into the letter grade converter.')
    num_grade = input()
    grade = int(num_grade)
    if grade < 60:
        print(" Letter grade is an 'F'!")
    elif grade > 59 and grade < 70:
        print(" Letter grade is a 'D'!")
    elif grade > 69 and grade < 80:
        print(" Letter grade is a 'C'!")
    elif grade > 79 and grade < 90:
        print(" Letter grade is a 'B'!")
    else:
        print("Congratulations! Your letter grade is an 'A'!")
    print('would you like to enter another grade into the converter? (Y/N)')
    prompt = input()
else:
    print("Thank you!")
```
